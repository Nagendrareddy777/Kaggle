# Line Charts

sns.lineplot - Line charts are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.

```python
#specifies the width and height of graph
fig=plt.figure(figsize=(10,4))

#lineplot is used for continuous data only in the feature
#x-axis represents the range of values in indexes
#y-axis represents the data from the feature
sns.lineplot(df["alcohol"])
```

```python
#you will also give multiple columns which gives you the comparison
sns.lineplot(x="alcohol",y="not_distracted",data=df)
```

```python
sns.lineplot(df[["alcohol","not_distracted"]])
```

```python
#passing entire dataset for lineplot
fig=plt.figure(figsize=(12,4))
sns.lineplot(df)
plt.xlabel("index")
plt.show()
```

# Heat Maps

sns.heatmap - Heatmaps can be used to find color-coded patterns in tables of numbers.

```python
df2=sns.load_dataset("car_crashes")
```

```python
#gives 5 rows randomly
df2.sample(5)
```

```python
df3=df2.drop("abbrev",axis=1)
```

```python
df3.head()
```

```python
#heatmaps are generally used to represent the relationship between the numerical data,it can also works for 
#categorical data by binning it to numerical data
plt.figure(figsize=(7,7))
sns.heatmap(data=df3,annot=True,cmap="Blues")
```

```python
#heat map for correlations
c=df2.corr(numeric_only=True)
sns.heatmap(data=c,annot=True,cmap="ocean")
```

# Reg plot

sns.regplot - Including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.

Reg plot is used to fit the regression line amoung the data,which makes us to easily analyze the relationship 
between the x axis and y axis

```python
sns.regplot(x="petal_length",y="petal_width",data=df1)
#by seeing the plot we observe while incresing the petal length ,the petal width also incresing
#having positive corelation
```

# lmplot

sns.lmplot - This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.

```python
#lm is used to fit the regression line amoung the group of data by hue column
sns.lmplot(x="petal_length",y="petal_width",data=df1,hue="species")
#sns.set_style("white")
```

```python
sns.lmplot(x="sepal_length",y="sepal_width",data=df1,hue="species")
#sns.set_context("poster")
#sns.set_style("dark")
```

# swarmplot

sns.swarmplot - Categorical scatter plots show the relationship between a continuous variable and a categorical variable.

its also known as categorical scatter plot

```python
sns.swarmplot(y="sepal_length",x="species",data=df1)
```

# Histograms

sns.histplot - Histograms show the distribution of a single numerical variable.

```python
sns.histplot(x="petal_length",data=df1,hue="species")
```

# Density plot

sns.kdeplot - KDE plots (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable 
(or two numerical variables).

kde - kernal density plot is used show the desity of the distribution

```python
sns.kdeplot(x="petal_length",data=df1,fill=True)
```

```python
sns.kdeplot(x="petal_length",data=df1,hue="species",shade=True)
```

```python
#distplot shows both the histogram and density plot(kdeplot)
sns.distplot(x=df1["petal_length"])
```

# 2D KDE plots(jointplot)

sns.jointplot - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.

```python
sns.jointplot(x="petal_length",y="petal_width",data=df1)
```

```python
sns.jointplot(x="petal_length",y="petal_width",data=df1,kind="kde")
```
